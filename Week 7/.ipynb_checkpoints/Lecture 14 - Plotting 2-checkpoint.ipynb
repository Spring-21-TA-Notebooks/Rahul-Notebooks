```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = range(300)
def gen_ys(obs):
    y = np.random.choice([-1, 0, 1], obs)
    return np.cumsum(y) # random walk
ys = [gen_ys(300) for _ in range(4)]
ys
```

### Multiple subplots in Matplotlib

```python
fig, ax = plt.subplots(2, 2)
flat_axis_list = [s for r in ax for s in r]
for s, y in zip(flat_axis_list, ys):
    s.plot(x, y)
```

```python
ax
```

### Other plot types

**Scatter**

```python
x = range(300)
y = gen_ys(300)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y, linestyle='', marker='.', markersize=1)
```

**Bar**

```python
x = [ 2,  4,  6,  8, 10]
y = [1., .2, .4, .7, .6]
```

```python
fig, ax = plt.subplots()
ax.bar(x, y)
```

**Box**

```python
x = [np.random.normal(10, 2, 100),
     np.random.normal(11, 1, 100),
     np.random.normal(15, 4, 100)]
```

```python
fig, ax = plt.subplots()
ax.boxplot(x);
```

### Seaborn

```python
df = pd.DataFrame(np.array(x).T, columns=['first', 'second', 'third'])
```

```python
df.head()
```

```python
df.shape
```

```python
sns.set()
sns.boxplot(data=df)
```

```python
#https://seaborn.pydata.org/examples/grouped_boxplot.html
tips = sns.load_dataset("tips")
tips.head()
```

```python
ax = sns.boxplot(x="day", y="total_bill",
            hue="smoker", palette=["m", "g"],
            data=tips)

sns.despine(offset=10, trim=True)
```

```python
sns.lineplot(x='total_bill', y='tip', data=tips)
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='smoker', data=tips);
```
