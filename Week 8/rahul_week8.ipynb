### Some code to help with the understanding - how to extract component text from a website

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = r'https://harris.uchicago.edu/academics/design-your-path/certificates/certificate-data-analytics'
response = requests.get(url)
soup= BeautifulSoup(response.text, 'lxml')
```
