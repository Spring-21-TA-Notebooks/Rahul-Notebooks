### Mapping

```python
# https://www.geeksforgeeks.org/python-map-function/
def addition(n):
    return n + n
  
# We double all numbers using map()
numbers = (1, 2, 3, 4)
result = map(addition, numbers)
print(list(result))
```

```python
def add_usa(state):
    return ''.join([state, ', USA'])
df_landfills['STATE'] = df_landfills['STATE'].map(add_usa)
df_landfills.head()
```
